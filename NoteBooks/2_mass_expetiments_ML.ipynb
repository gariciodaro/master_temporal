```python
def run_experiment(experiment):
    #experiment="_entropy_delta_theta"
    data_co="/home/gari/Desktop/master_tesis_v3/Data/Datasets_"
    data_origin=data_co+experiment+"/prepared_train_data_set.file"
    
    data_test_path=data_co+experiment+"/prepared_test_data_set.file"


    note_exp=experiment

    data=OFH.load_object(data_origin)
    score_1=CoreML.core_diagnosis(data_origin=data,
                      data_origin_string="object_dataframe",
                      name="/eeg_"+experiment,
                      string_metric="roc_auc_score",
                      number_of_clusters=200,
                      hiper_tunning_iters=200,
                      futher_op_boo=True,
                      balance_train=False,
                      clustering=True,
                      polynomial=True,
                      test_size=0.2,
                      path_to_test=data_test_path,
                      note_exp=note_exp)
    print("EEG",score_1) 
```
