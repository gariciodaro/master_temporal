```python
def join_all_exp_with_subject_id(list_id):
    """
    returns a dataframe of all predictions 
    by experiment and by test subject from
    the excel file.
    """
    for index,each_id in enumerate(list_id):
        df=pd.read_excel('./EXPERIMENT_CONTROL.xlsx', sheet_name=each_id)
        df["test_patient_id"]=df["test_patient_id"].str.strip()
        df=df.set_index("test_patient_id")
        if index==0:
            df_holder=df
        else:
            df_holder=df_holder.join(df[["Predicted_Target","ajusted_decision"]],rsuffix="_"+each_id)
            
    df_holder=df_holder.rename(columns={"Predicted_Target":"Predicted_Target_ID_1",
                                        "ajusted_decision":"ajusted_decision_ID_1"})
    return df_holder
            
    
```

<div style="height:300;background-color:#021A36;color:white;border-radius: 15px;padding:10px">
    <p>Let's exclude problematic subjects from the test set. I defined them as subjects who were misclassified above     the average times 1.5  within thier respective  group: "Combined Type of ADHD," "Inattentive Type of ADHD," and "No Diagnosis Given." Note that this exclusion was not performed for the ADHD-hyperactive / impulsive type, as we only had two subjects in the test set. Furthermore, they were correctly classified most of the time.</p>
    <p>Some of the excluded subject were systimatically missclasified; this means most of the classifiers concluded that the patients had ADHD but the biobank label was healthy, I will explore this cases later.</p>
</div>

```python
excluded_test_subjects=[]

for label,df in test_subject_analysis.groupby("DX_01"):
    if label!="ADHD-Hyperactive/Impulsive Type":
        miss=df.sum(axis=1)
        avg_miss=miss.mean()
        miss_df=pd.DataFrame(miss.sort_values(ascending=False),columns=["times_missclassified"])
        times_avg=1.48
        exclude=list(miss_df[miss_df["times_missclassified"]>avg_miss*times_avg].index)
        print("Total excluded in:",label, len(exclude), "out of: ",len(miss_df), "criterium:",avg_miss*times_avg)
        excluded_test_subjects=excluded_test_subjects+exclude
        fig=plt.figure(figsize=(15,7))
        ax=fig.add_subplot(1,1,1)
        #colors=['#591303' for each in range(len(miss_df)-len(exclude))]
        #colors2=['#035AA6' for each in range(len(exclude))]
        
        miss_df.plot(kind="bar",stacked=True,
             alpha=0.9,ax=ax,color='#022340')
        
        ax.axhline(y=avg_miss*times_avg, color='r', linestyle='--',label="Threshold: "+str(int(avg_miss*times_avg)))
        
        plt.legend(loc=0)
        
        ax.set_title(label+", total subjects:"+str(len(miss))+ ", Total clf:"+str(len(df.columns))  )
        
```

<div style="height:200;background-color:#021A36;color:white;border-radius: 15px;padding:10px">
    <p>Let us recalculate the precision of each classifier without the problematic subjects. Also, let us calculate the t-test with the fallowing null hypothesis</p>
    $H_0:$ The average probality prediction per Group (Heathy vs ADHD) is equal. 
    <p>With the recalculated presicion and the p-value of that null hypothesis, we can rank the modeling procedure, and select the 3 best</p>
</div>

```python
from sklearn.metrics import precision_score,f1_score

ajusted_des_col=["ajusted_decision_ID_"+str(col)for col in range(1,47)]
#ajusted_des_col.append("ajusted_decision_ID_40")
Predicted_Target_col=["Predicted_Target_ID_"+str(col)for col in range(1,47)]
#Predicted_Target_col.append("Predicted_Target_ID_40")

scores={}
f1_scores={}
t_test={}
mean_HEALTHY={}
mean_ADHD={}
for tuple_interes in zip(ajusted_des_col,Predicted_Target_col):
    each_ajusted_col=tuple_interes[0]
    each_predicted_target=tuple_interes[1]

    experiemnt_id="_".join(each_ajusted_col.split("_")[-2:])
    scores[experiemnt_id]=precision_score(df_expe_no_outli.target,df_expe_no_outli[each_ajusted_col])
    f1_scores[experiemnt_id]=f1_score(df_expe_no_outli.target,df_expe_no_outli[each_ajusted_col])
    
    ADHD=df_expe_no_outli[df_expe_no_outli["target"]==1][[each_predicted_target]]
    HEALTHY=df_expe_no_outli[df_expe_no_outli["target"]==0][[each_predicted_target]]
    
    t_test[experiemnt_id]=ttest_ind(ADHD,HEALTHY,equal_var=False)[1][0]
    
    t_test[experiemnt_id]=ttest_ind(ADHD,HEALTHY,equal_var=False)[1][0]
    mean_HEALTHY[experiemnt_id]=HEALTHY.mean()[0]
    mean_ADHD[experiemnt_id]=ADHD.mean()[0]


precision_score_df=pd.DataFrame.from_dict(scores, orient='index',columns=["new_precision"])
f1_score_df=pd.DataFrame.from_dict(f1_scores, orient='index',columns=["new_f1_score"])
t_test_t_df=pd.DataFrame.from_dict(t_test, orient='index',columns=["p_value"])
mean_HEALTHY_df=pd.DataFrame.from_dict(mean_HEALTHY, orient='index',columns=["mean_HEALTHY"])
mean_ADHD_df=pd.DataFrame.from_dict(mean_ADHD, orient='index',columns=["mean_ADHD"])
summary_models=precision_score_df.join(f1_score_df).join(t_test_t_df).join(mean_HEALTHY_df).join(mean_ADHD_df)
```

```python
ranked_mp=summary_models.sort_values(by=["p_value","new_f1_score"],ascending=[True,False])
ranked_mp.join(experiments_summary[["pre_processing",
                                    "Features",
                                    "Machine_learning_technique",
                                    "Transformation",
                                    "train score","precision"]])
```

```python
def execute_test(by_c,by_2c,df_focus,mode,col):
    labels=[]
    skip_index=0
    for label, df in df_focus[[by_2c,by_c]].groupby(by_c):
        #print(df,label)
        for index,element in enumerate(df_focus[[by_2c,by_c]].groupby(by_c)):
            label2,df2=element[0],element[1]
            if label!=label2 and index>=skip_index:
                if mode=="mean":
                    print("t-test on:",label,label2)
                    t_test=ttest_ind(df[col],df2[col],equal_var=True)
                    print(t_test)
                    print("mean prediction for :",label,df[col].mean())
                    print("mean prediction for :",label2,df2[col].mean())
                    print("*"*100)
                if mode=="var":
                    print("bartlett on:",label,label2)
                    t_test=bartlett(df[col],df2[col])
                    print(t_test)
                    print("variance prediction for :",label,df[col].var())
                    print("variance prediction for :",label2,df2[col].var())
                    print("*"*100)
        skip_index=skip_index+1


def generate_plots(df,title,by_c="Sex",by_2c="Predicted_Target",size=(17,9),rot=90):
    """
    Generete distributions plot
    for visual exploration.
    """
    fig=plt.figure(figsize=size)

    ax0=fig.add_subplot(2,2,1)
    ax1=fig.add_subplot(2,2,2)

    ax2=fig.add_subplot(2,2,3)
    ax3=fig.add_subplot(2,2,4)
    
    colors_list=sns.color_palette("RdBu", len(df[by_c].unique())+3)
    #colors_list.reverse()

    df[by_c].value_counts().plot(kind="pie",
                                  autopct='%1.1f%%',
                                  startangle=rot,
                                  shadow=True,
                                  ax=ax0,colors=colors_list)
    
    #ax0.axis('equal') 
    #ax0.legend(labels=df[by_c].value_counts().index, loc='upper left')
    ax0.set_ylabel('')
    
    
    df[[by_2c,by_c]].boxplot(by=by_c,ax=ax3, rot=90)
    ax3.set_xlabel('')
    
    
    df[by_c].value_counts().plot(kind="bar",ax=ax2,color=colors_list)
    #df[by_c].value_counts().plot(kind="bar",ax=ax2)
    
    
    i=0
    for label, df_in in df[[by_2c,by_c]].groupby(by_c):
        try:
            #df_in[by_2c].plot(kind="kde", ax=ax1, label=label)
            if(label=="ADHD-Hyperactive/Impulsive Type"):
                #sns.distplot(df_in[by_2c],bins=10 ,ax=ax1, label=label,kde=False,hist_kws={"alpha": 0.5},color="black")
                pass
            elif(label=="No Diagnosis Given"):
                sns.distplot(df_in[by_2c],bins=10 ,ax=ax1, label=label,
                             kde=False,hist_kws={"alpha": 0.5},color="black")
            else:
                sns.distplot(df_in[by_2c],bins=10 ,ax=ax1, label=label,
                             kde=False,hist_kws={"alpha": 0.4},color=colors_list[i])
        except:
            pass
        i=i+1
    
    #sns.distplot(df[[by_2c,by_c]],bins=10,kde=False,hist_kws={"alpha": 0.5})
    
    execute_test(by_c=by_c,by_2c=by_2c,df_focus=df,mode="mean",col=by_2c)
    execute_test(by_c=by_c,by_2c=by_2c,df_focus=df,mode="var",col=by_2c)
    
    ax1.legend()
    #ax1.set_title("Subjects Gender %. Age < 18")
    ax0.set_title("distribution by "+by_c+" total: "+str(len(df)))
    #plt.title(title)
    fig.suptitle(title, fontsize=14)
    #plt.legend(['u','v','u+v'])
```
