```python
def get_predictions(learn,path,df_test):
    """
    takes df_test dataframe
    makes predictions
    """
    prediction=[]
    for each_image in df_test.name:
        #print(each_image)
        img = open_image(path+"test_N/"+each_image)
        p=learn.predict(img)
        tensor = p[2]
        single_pred=tensor.cpu().detach().numpy()[0]
        prediction.append(single_pred)

    pred_data=pd.DataFrame(prediction,columns=["Predicted_Target"])
    return pred_data

def get_max_and_mean_acc(df_predictions,df_fast_test,mode):

    df_subject=df_test_identity.join(df_predictions).join(df_fast_test[["target"]])

    if mode=="max":
        df_subject_pred=df_subject.groupby("test_patient_id").max()
    if mode=="mean":
        df_subject_pred=df_subject.groupby("test_patient_id").mean()
    
    
    df_subject_pred.loc[df_subject_pred.Predicted_Target > 0.5, "ajusted_decision"] = 0
    df_subject_pred=df_subject_pred.fillna(1)
    ajusted_acc=accuracy_score(df_subject_pred.target, df_subject_pred.ajusted_decision)
    #if mode=="mean":
        #print(mode+" "+"*"*100)
        #print(df_subject_pred)
    return ajusted_acc,df_subject_pred

path_identity="/home/gari/Desktop/master_tesis/adhd_detector_v2/Data/"
df_train_identity=OFH.load_object(path_identity+"df_train_identity.file")
df_test_identity=OFH.load_object(path_identity+"df_test_identity.file")
```

```python
np.random.seed(42)


experiment="spectrum"

string_models=["resnet18","resnet50","alexnet","squeezenet1_0","densenet169"]
string_model=string_models[4]


root_path_exp="/home/gari/Desktop/master_tesis/adhd_fast_ai/Data_fast_ai/"+experiment+"/"


print("*"*100)
print("string_model= ",string_model)
bs=4
path=root_path_exp
string_model=string_model
pretrained=True
number_epochs=5
ps=0.00
normalize=True

dic_models={"resnet18":models.resnet18,
            "resnet50":models.resnet50,
            "alexnet":models.alexnet,
            "squeezenet1_0":models.squeezenet1_0,
            "densenet169":models.densenet169
            }

if normalize:
    data = ImageDataBunch.from_folder(path,bs=bs).normalize(imagenet_stats)
else:
    data = ImageDataBunch.from_folder(path,bs=bs)

#data.show_batch(rows=3, figsize=(7,6))

#print(data)
#print(data.classes, data.c, len(data.train_ds), len(data.valid_ds))

#learn = cnn_learner(data,dic_models.get(string_model), metrics=metrics=[accuracy, AUROC()],pretrained=True,ps=ps)
learn = cnn_learner(data,dic_models.get(string_model), metrics=[accuracy, AUROC()],pretrained=pretrained,ps=ps)
#learn.unfreeze()
learn.fit_one_cycle(number_epochs)

interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
plt.savefig('./Data_fast_ai/'+experiment+'/confusion_matrix_'+string_model+'.png')


df_fast_test=OFH.load_object(path+"test_fast_ai.file")


df_predictions=get_predictions(learn,path,df_fast_test)

#max_score=get_max_and_mean_acc(df_predictions,df_fast_test,"max")
mean_score,df_test_results=get_max_and_mean_acc(df_predictions,df_fast_test,"mean")

if string_model=="resnet18":
    df_holder_results=df_test_results
    OFH.save_object('./Data_fast_ai/'+experiment+'/df_holder_results.file',df_holder_results)

else:
    df_holder_results=OFH.load_object('./Data_fast_ai/'+experiment+'/df_holder_results.file')
    df_holder=df_holder_results.join(df_test_results,lsuffix="_"+string_model)
    OFH.save_object('./Data_fast_ai/'+experiment+'/df_holder_results.file',df_holder)


#learn.destroy()
#torch.cuda.empty_cache() 
#print("ajusted max score acc: ",max_score)
print("ajusted mean score acc: ",mean_score)
print("+"*100)
```

```python
def pretty_size(size):
	"""Pretty prints a torch.Size object"""
	assert(isinstance(size, torch.Size))
	return " × ".join(map(str, size))

def dump_tensors(gpu_only=True):
	"""Prints a list of the Tensors being tracked by the garbage collector."""
	import gc
	total_size = 0
	for obj in gc.get_objects():
		try:
			if torch.is_tensor(obj):
				if not gpu_only or obj.is_cuda:
					print("%s:%s%s %s" % (type(obj).__name__, 
										  " GPU" if obj.is_cuda else "",
										  " pinned" if obj.is_pinned else "",
										  pretty_size(obj.size())))
					total_size += obj.numel()
			elif hasattr(obj, "data") and torch.is_tensor(obj.data):
				if not gpu_only or obj.is_cuda:
					print("%s → %s:%s%s%s%s %s" % (type(obj).__name__, 
												   type(obj.data).__name__, 
												   " GPU" if obj.is_cuda else "",
												   " pinned" if obj.data.is_pinned else "",
												   " grad" if obj.requires_grad else "", 
												   " volatile" if obj.volatile else "",
												   pretty_size(obj.data.size())))
					total_size += obj.data.numel()
		except Exception as e:
			pass        
	print("Total size:", total_size)
```
